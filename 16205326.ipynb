# Visualisation 1 2 and 3 dataset manipulation



```python
#importing libraries 
import csv
import pandas as pd
```

## Reading the intial dataset

```python
data=pd.read_csv("../datasets/initial_dataset.csv")
data
```

## Cleaning the initial dataset
Changing all non sccessful statuses to failed, as there is no need for details

```python
cleaned_data=data[['category','state','usd_goal_real','country','launched']]#.to_csv("../datasets/visualisation1.csv")


def remove_all_not_successful(row):
    if row.state=='successful':
        return 'successful'
    else:
        return 'fail'

cleaned_data['state']=cleaned_data.apply(lambda row: remove_all_not_successful(row), axis=1)
print(cleaned_data[169:170])
cleaned_data.loc[cleaned_data['country'] == 'N,0"', 'country'] = "Not Specified"

cleaned_data[0:10]

```

## Cleaning and aggregating data for the visualisation 1

```python
vis1=cleaned_data[['category', 'state']]
vis1['count']=0
 

vis1 = vis1.groupby(['category', 'state']).count()
```

saving data for the visualisation 1

```python
vis1.to_csv("../datasets/vis1.csv")
print(vis1)
```

## Cleaning and aggregating and saving data for the visualisation 2

```python
vis2=cleaned_data[['category', 'country', 'usd_goal_real']]
vis2 = vis2.groupby(['category', 'country']).median()
vis2.to_csv("../datasets/vis2.csv")
print(vis2)
```

## Cleaning and aggregating and saving data for the visualisation 3

```python
vis3=cleaned_data[['launched']]


datetimes = pd.to_datetime(vis3['launched'])
vis3['year'] = datetimes.dt.year 
vis3['month'] = datetimes.dt.month

# creating sorter value
vis3['sorter']= vis3['year']*100+vis3['month']


#creating string representation of time
vis3['year']=vis3['year'].astype(str)
vis3['month']=vis3['month'].astype(str)
vis3['launched'] = vis3['year']+"-"+vis3['month']


vis3=vis3[['launched','sorter']]

vis3 = vis3.sort_values('sorter')

vis3.to_csv("../datasets/vis3.csv")
print(vis3)

```
